# Import libraries and read data

```python
import plotly.express as px
import pandas as pd
import sweetviz as sv
import klib
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('../../dataset/proper_dataset1.csv')
```

```python
data.head()
```

# Analyze using sweetviz

```python
v = sv.analyze(data)
```

```python
v.show_html('Basic_eda.html', layout='vertical')
```

# Analyze using klib

## correlation plot

```python
# positive  correlation plot

plot = klib.corr_plot(data, split='pos')
fig = plot.get_figure()
fig.savefig("positive_corr.jpg")
```

```python
# negative  correlation plot

plot = klib.corr_plot(data, split='neg')
fig = plot.get_figure()
fig.savefig("negative_corr.jpg")
```

```python
# correlation with target feature

plot = klib.corr_plot(data, target='fertility')
fig = plot.get_figure()
fig.savefig("target_corr.jpg")
```

# Analyze using plotly

```python
# scatter plot N vs ph

fig = px.scatter(data, x='N', y='ph', color='fertility')
fig.show()
fig.write_image('scatterplot_N_vs_ph.jpg')
```

```python
# scatter plot N vs oc

fig = px.scatter(data, x='N', y='oc', color='fertility')
fig.show()
fig.write_image('scatterplot_N_vs_oc.jpg')
```

```python
# px.histogram(data, x='N', marginal='rug')
```

```python
# boxplot fertility vs N

fig = px.box(data, y='N', x='fertility')
fig.update_layout(height=600)
fig.show()
fig.write_image('boxplot_fertility_vs_N.jpg')
```

```python
# box plot fertility vs K

fig = px.box(data, y='K', x='fertility')
fig.update_layout(height=500)
fig.show()
fig.write_image('boxplot_fertility_vs_K.jpg')
```

```python
# boxplot fertility vs fe

fig = px.box(data, y='fe', x='fertility')
fig.update_layout(height=500)
fig.show()
fig.write_image('boxplot_fertility_vs_fe.jpg')
```
