# Import libraries and read data

```python
import plotly.express as px
import pandas as pd
import sweetviz as sv
import klib
```

```python
data = pd.read_csv('./proper_dataset1.csv')
```

```python
data.head()
```

# Analyze using sweetviz

```python
v = sv.analyze(data)
```

```python
v.show_html('Basic_eda.html', layout='vertical')
```

# Analyze using klib

## correlation plot

```python
# positive  correlation plot

klib.corr_plot(data, split='pos')
```

```python
# negative  correlation plot

klib.corr_plot(data, split='neg')
```

```python
# correlation with target feature

klib.corr_plot(data, target='fertility')
```

## dist plot

```python
# distribution plot showing the stats of all independent features
df = data.drop('fertility', axis=1)

klib.dist_plot(df)
```

# Analyze using plotly

```python
# scatter plot N vs ph

fig = px.scatter(data, x='N', y='ph', color='fertility')
fig.show()
```

```python
# scatter plot N vs oc

fig = px.scatter(data, x='N', y='oc', color='fertility')
fig.show()
```

```python
# px.histogram(data, x='N', marginal='rug')
```

```python
# boxplot fertility vs N

fig = px.box(data, y='N', x='fertility')
fig.update_layout(height=600)
fig.show()
```

```python
# box plot fertility vs K

fig = px.box(data, y='K', x='fertility')
fig.update_layout(height=500)
fig.show()
```

```python
# boxplot fertility vs fe

fig = px.box(data, y='fe', x='fertility')
fig.update_layout(height=500)
fig.show()
```
